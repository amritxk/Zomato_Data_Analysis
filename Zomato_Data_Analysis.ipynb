```python
# Check in how many countries Zomato is working in
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_val= final_df.Country.value_counts().values
```

```python
#Top 3 countries which uses zomato

plt.pie(country_val[:3] , labels=country_names[:3] , autopct='%1.2f%%')
```

It is clear that most numbers of transcations are happening in india which is obvious because zomato is a india based company. After India it is US and then UK.

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns ={0:'Rating count'})
ratings
```

Observations
1. When ratings between 4.5 to 4.9 --> Excellent
2. When ratings between 4.0 to 4.4 --> Very good
3. When rating between 3.5 to 3.9 --> good
4. when rating between 3.0 to 3.4 --> average
5. when rating between 2.5 to 2.9 --> average
6. when rating between 1.8 to 2.4 --> poor


```python
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color' , data=ratings , palette=['grey','red','orange','yellow','cyan','green'])
```

Observations
1. Not rated count is very high
2. Maximum no of ratings are between 2.5 to 3.4
